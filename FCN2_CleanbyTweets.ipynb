# Check by Msg

```python
import pandas as pd

df = pd.read_csv('/Users/wenlanzhang/Downloads/PhD_UCL/Data/Twitter/Nairobi_1922/Sum/Summary.csv', lineterminator='\n')
df
```

## Tweets shorter than 3 words

```python
df['Len'] = df['text'].apply(lambda x: len(x.split()))
df = df[df['Len'] > 3]
```

## Same user repeated tweets

```python
df['IDFreq'] = df.groupby('userid')['userid'].transform('count')
df['MsgFreq'] = df.groupby('text')['text'].transform('count')
df = df.sort_values('MsgFreq')

df['IdMsgFreq'] = 0

df
```

### Select valid msg freq users 

```python
# Select valid msg freq users for append later

dfmsgV = df[df['MsgFreq'] <= 2]  # 完全一样的信息重复两次以下
dfmsgV
```

### Select unvalid msg freq users 

```python
# Select high freq msg

dfmsg = df[df['MsgFreq'] > 2]
dfmsg.head(30)
```

```python
# Select users with repeated tweets
Uid = list(set(dfmsg.userid.tolist()))
print(Uid)
print(len(Uid))
```

```python
# Append tweets if the message frequency is lower than 3

for id in Uid:
# id = 579141934
    dftemp = dfmsg[dfmsg['userid'] == id]
    dftemp['IdMsgFreq'] = dftemp.groupby('text')['text'].transform('count')
    if (dftemp['IdMsgFreq'] > 3).any():
        continue
    else:
        dfmsgV = dfmsgV.append(dftemp)
dfmsgV
# 6min
```

```python
# dfmsgV.drop(["IDFreq", "MsgFreq", 'IdMsgFreq', 'Len'], axis=1, inplace=True)
dfmsgV["created_at"] = pd.to_datetime(dfmsgV["created_at"])
dfmsgV = dfmsgV.sort_values(by=['created_at'])
dfmsgV
```

```python
dfmsgV.to_csv('/Users/wenlanzhang/Downloads/PhD_UCL/Data/Twitter/Nairobi_1922/Sum/' + 'Clean1Tweets.csv', index = False)
```
